```python
dfp=pd.read_csv('LD1_P201601~201712.csv', parse_dates=[['DATE', 'TIME']]) 
dfq=pd.read_csv('LD2_QT201601~201712.csv', header = 1, parse_dates=[['DATE', 'TIME']]) 

dfs1_mes=pd.read_csv('Res_comb_S1_mes.csv',parse_dates=[['DATE', 'TIME']]) 
dfs1_mes_tz=pd.read_csv('Res_comb_S1_mes_tz_Jp.csv',parse_dates=[['DATE', 'TIME']])
dfs2_mes=pd.read_csv('Res_comb_S2_mes.csv',parse_dates=[['DATE', 'TIME']]) 
dfs2_mes_tz=pd.read_csv('Res_comb_S2_mes_tz_Jp.csv',parse_dates=[['DATE', 'TIME']])

dfs1_mes2017_tz=pd.read_csv('Res_comb_S1_mes_2017_tz_Jp.csv',parse_dates=[['DATE', 'TIME']])
dfs2_mes2017_tz=pd.read_csv('Res_comb_S2_mes_2017_tz_Jp.csv',parse_dates=[['DATE', 'TIME']])

dfq.head()
# dfp.head()
# dfs1_mes_tz.head()
# print(len(dfp), len(dfq))
# print(len(dfs1_mes), len(dfs2_mes), len(dfs1_mes_tz), len(dfs2_mes_tz))

# list(dfq)
# list(dfp)
# list(dfs1_mes)
```

Merge, join, and concatenate

https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

https://stackoverflow.com/questions/36292959/pandas-merge-data-frames-on-datetime-index


```python
dfp.set_index(pd.to_datetime(dfp['DATE_TIME']), inplace=True)
dfq.set_index(pd.to_datetime(dfq['DATE_TIME']), inplace=True)

#### dfp.join(dfq, how='right')
dfpq=pd.merge(dfp,dfq, how='inner', on='DATE_TIME')
dfpq.head()
```

```python
dfs1_mes2017_tz.set_index(pd.to_datetime(dfs1_mes2017_tz['DATE_TIME']), inplace=True)
dfs2_mes2017_tz.set_index(pd.to_datetime(dfs2_mes2017_tz['DATE_TIME']), inplace=True)

dfpq_s1_2017mes=pd.merge(dfpq,dfs1_mes2017_tz, how='inner', on='DATE_TIME')
dfpq_s1_2017mes.head()

dfpq_s2_2017mes=pd.merge(dfpq,dfs2_mes2017_tz, how='inner', on='DATE_TIME')
dfpq_s2_2017mes.head()
```

```python
print(len(dfpq_s1_2017mes), len(dfpq_s2_2017mes))
```

NOTE: S1 and S2 can be aslo merged together with the power and measurements (that could increas the training power for the forecasting mode).

```python
dfpq_s1_s2_2017mes=pd.merge(dfpq_s1_2017mes,dfs2_mes2017_tz, how='inner', on='DATE_TIME')
dfpq_s1_s2_2017mes.head()
```

```python
print(len(dfpq_s1_s2_2017mes)) #number of rows
print(len(dfpq_s1_s2_2017mes.columns))  #number of columns
```
